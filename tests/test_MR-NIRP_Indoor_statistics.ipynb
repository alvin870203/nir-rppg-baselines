```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rppg_signals_diff = np.load('../rppg_signals_diff.npy')

np.info(rppg_signals_diff)
print(f"{rppg_signals_diff.mean()=}, {rppg_signals_diff.std()=}")

plt.figure(figsize=(20, 6))
plt.plot(rppg_signals_diff/ rppg_signals_diff.std(), 'o', markersize=2)
plt.show()

plt.figure(figsize=(20, 6))
plt.hist(rppg_signals_diff/ rppg_signals_diff.std(), bins=100)
plt.show()

plt.figure(figsize=(20, 6))
plt.plot(rppg_signals_diff, 'o', markersize=2)
plt.show()

plt.figure(figsize=(20, 6))
plt.hist(rppg_signals_diff, bins=100)
plt.show()
```
